```python
# Q1. Explain the significance of Python keywords and provide examples of five keywords

"""
Python keywords are reserved words that have specific meanings and purposes within the language.
These keywords cannot be used as variable names, function names, or any other identifiers.

(1) Value Keywords:
True, False, and None:
These represent boolean truth values and the absence of a value, respectively.
Example:
"""
flag = True
result = None
print(result)

"""
(2) Operator Keywords:
and, or, not, in, and is:
These are used for logical operations, membership checks, and identity comparisons.
Example:
"""
x = 10
y = 20
if x > 0 and y < 30:
    print("Both conditions are true")
    
"""
(3) Control Flow Keywords:
if, elif, and else:
Used for conditional branching.
Example:
"""
age = 18
if age >= 18:
    print("You are an adult")
else:
    print("You are not yet an adult")

"""
(4) Iteration Keywords:
for and while:
Used for loops.
Example:
"""
for i in range(5):
    print(i)
    
"""
(5) Structure Keywords:
def, class, with, as, and lambda:
Used for defining functions, classes, context managers, and anonymous functions (lambda expressions).
Example:
"""
def greet(name):
    return f"Hello, {name}!"

greet("Vinay")
```

```python
# Q2. Describe the rules for defining identifiers in Python and provide an example

"""
In Python, identifiers are user-defined names given to variables, functions, classes, modules, and other objects.
These names play a crucial role in making code more readable and understandable.

Keywords:
Keywords (reserved words) cannot be used as identifier names. Attempting to do so will result in a SyntaxError.
Examples of keywords include if, else, for, while, and def.

Valid Characters:
Identifiers can contain:
Letters (both lowercase and uppercase): a-z, A-Z
Digits: 0-9
Underscore: _
They cannot begin with a digit.
For example:
Valid: my_variable, count_123
Invalid: 10test, 888

Starting with an Underscore:
Identifiers can start with an underscore (_).
For example: _test

Case Sensitivity:
Identifier names are case-sensitive.
abc and ABC are considered different identifiers.

Length Limit:
There is no strict limit on the length of an identifier, but excessively long names are discouraged.
Keep identifiers concise for readability.

Examples :- 

Valid Identifiers:
    ab10c: Contains only letters and numbers.
    abc_DE: Contains letters, numbers, and underscores.
    _: Surprisingly, a single underscore is a valid identifier.
    _abc: Identifiers can start with an underscore.

Invalid Identifiers:
    99: Cannot be only digits.
    9abc: Cannot start with a number.
    x+y: Only underscores are allowed as special characters.
    for: Reserved keyword in Python.
"""
```
